torch.no_grad()

What it does:
A context manager that tells PyTorch not to build a computation graph for the operations inside its block.

optimizer.zero_grad() or model.zero_grad()

What it does:
Sets the .grad attributes of all model parameters to zero.
(Since PyTorch accumulates gradients by default after every .backward() call, you must clear them at the start of each training step.)

```python
import torch
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

## Define the model

```python
class MySimpleNN():

  def __init__(self,X):
    self.weights=torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
    self.bias=torch.zeros(1,dtype=torch.float64,requires_grad=True)

  def forward(self,X):
    z=torch.matmul(X,self.weights)+self.bias
    y_pred=torch.sigmoid(z)
    return y_pred

  def loss_function(self,y_pred,y_train_tensor):
    epsilon=1e-7
    y_pred=torch.clamp(y_pred,epsilon,1-epsilon)
    loss=-(y_train_tensor*torch.log(y_pred)+(1-y_train_tensor)*torch.log(1-y_pred)).mean()
    return loss

```

```python
learning_rate=0.1
epochs=25
```

## Training Pipeline

```python
# create model
model=MySimpleNN(X_train_tensor)

# define loop
for epoch in range(epochs):

  # forward pass
  y_pred=model.forward(X_train_tensor)

  # calculate loss
  loss=model.loss_function(y_pred,y_train_tensor)
  #print(f'epoch:{epoch},loss:{loss.item()}')

  # backward pass
  loss.backward()

  #parameters update
  # NO computation graph built
  with torch.no_grad():
    model.weights -= learning_rate*model.weights.grad
    model.bias -= learning_rate*model.bias.grad

  # zero gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f'epoch:{epoch},loss:{loss.item()}')
```

```python
# Applying on Dataset above

from torch.utils.data import Dataset,DataLoader

class CustomDataset(Dataset):
  def __init__(self,features,labels):
    self.fetaures=features
    self.labels=labels

  def __len__(self):
    return len(self.labels)

  def __getitem__(self,idx):
    return self.fetaures[idx],self.labels[idx]
```

```python
train_dataset=CustomDataset(X_train_tensor,y_train_tensor)
test_dataset=CustomDataset(X_test_tensor,y_test_tensor)
```

```python
train_dataset[0]
```

```python
train_data_loader=DataLoader(train_dataset,batch_size=32,shuffle=True)
test_data_loader=DataLoader(test_dataset,batch_size=32,shuffle=True)
```

```python
import torch.nn as nn

class MySimpleNN(nn.Module):

  def __init__(self,num_features):
    super().__init__()
    self.linear=nn.Linear(num_features,1)
    self.sigmoid=nn.Sigmoid()


  def forward(self,features):
    out=self.linear(features)
    out=self.sigmoid(out)
    return out
```

```python
learning_rate=0.1
epochs=25
```

```python
model=MySimpleNN(num_features=X_train_tensor.shape[1])
```

```python
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)
loss_fn=nn.BCELoss()
```

```python
for epoch in range(epochs):

  for batch_fetaures,batch_labels in train_data_loader:

    y_pred=model(batch_fetaures)

    loss=loss_fn(y_pred,batch_labels.view(-1,1))

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

  print(f"Epoch: {epoch} Loss: {loss.item()}")
```

```python
model.eval()
```

```python
accuracy_list=[]

with torch.no_grad():
  for batch_fetaures,batch_labels in test_data_loader:

    y_pred=model(batch_fetaures)

    y_pred=(y_pred>0.8).float()

    batch_accuracy=(y_pred.view(-1)==batch_labels).float().mean().item()
    accuracy_list.append(batch_accuracy)


overall_accuracy=sum(accuracy_list)/len(accuracy_list)
print(f"Overall Accuracy: {overall_accuracy}")
```
